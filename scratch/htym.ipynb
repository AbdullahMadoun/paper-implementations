```python
import numpy as np 
import matplotlib.pyplot as plt 
import math
```

```python
def roofline(B, D, F):
    total_flops = 2*B*D*F
    time_comms = (2*B*D + D*F + 2 * B* F ) / 8.2e11
    time_comps = total_flops / 1.97e14
    time_total = np.maximum(time_comms, time_comps)
    return total_flops / time_total
```

```python
batches = np.arange(0, 1024, 1)
```

```python
plt.figure(figsize = (10,4))
plt.plot(batches,roofline(batches, 1024, 1024), c ="b")
plt.plot(batches,roofline(batches, 4096, 4096), c = "r")

plt.show()
```
