```python
weight = 0.7
bias = 0.3

start =-10
end = 10
step = 0.2
X = torch.arange(start , end, step).unsqueeze(dim=1)
y = weight * X*X*X*X + bias 

 
```

```python
def plot_prediction(train_data=X_train,
                    train_labels=y_train,
                    test_data=X_test,
                    test_labels=y_test,
                    predictions = None
): 
    """ 
    Plot data
    """ 
    plt.figure(figsize = (10,7))
    plt.scatter(train_data, train_labels, c="r", s=4, label ="Training data")
    plt.scatter(test_data, test_labels, c="b", s=4, label ="Test data")
    if predictions is not None: 
        plt.scatter(test_data,predictions, c = "g", s=4, label ="Predictions")
    plt.legend(prop=({"size":14}))
```

```python
class QuadraticRegressionModel(nn.Module): 
    def __init__(self):
        super().__init__()
        self.weights = nn.Parameter(torch.randn(1, 
                                    requires_grad= True,
                                   dtype = torch.float)
                                   )
        self.bias = nn.Parameter(torch.randn(1, 
                                    requires_grad= True,
                                   dtype = torch.float )
                                   )
    def forward(self, x: torch.Tensor) -> torch.Tensor: 
            return self.weights*x*x + self.bias
        
```

```python
torch.manual_seed(42)
model_0 = QuadraticRegressionModel()
list(model_0.parameters())
```

```python
loss_fn = nn.MSELoss()
optimizer = torch.optim.AdamW(
    lr = 0.01, 
    params = model_0.parameters(),
    
)
```

```python
import torch

def generate_X(dim=100, batch_size=None, num_steps=10):
    """
    Generator yielding X tensors.
    
    Parameters:
    - dim (int): Feature dimension (matches your dim1=100).
    - batch_size (int, optional): 
        - If None: Yields unbatched tensors of shape (dim, 1)
        - If integer (e.g., 32): Yields batched tensors of shape (batch_size, dim, 1)
    - num_steps (int): Number of batches/samples to generate.
    """
    for _ in range(num_steps):
        if batch_size is None:
            yield torch.randn(dim, 1)
        else:
            yield torch.randn(batch_size, dim, 1)

# --- Example Usage ---

# 1. Unbatched Generator (yields shape: [100, 1])
single_gen = generate_X(dim=100, batch_size=None, num_steps=3)
for X_single in single_gen:
    print("Unbatched X shape:", X_single.shape)

# 2. Batched Generator (yields shape: [batch_size, 100, 1])
batch_gen = generate_X(dim=100, batch_size=32, num_steps=3)
for X_batched in batch_gen:
    print("Batched X shape:", X_batched.shape)
```

```python
import torch

def generate_data(dim_x=100, dim_y=5, batch_size=None, num_steps=10, target_type="regression"):
    """
    Generator yielding (X, Y) tensor pairs.
    
    Parameters:
    - dim_x (int): Feature dimension of X (matches dim1=100).
    - dim_y (int): Output/target dimension of Y (matches the final layer output).
    - batch_size (int, optional): 
        - If None: Yields unbatched tensors: X shape (dim_x, 1), Y shape (dim_y, 1) or scalar.
        - If integer: Yields batched tensors: X shape (batch_size, dim_x, 1), Y shape (batch_size, dim_y, 1) or (batch_size,).
    - num_steps (int): Number of batches/samples to generate.
    - target_type (str): "regression" (continuous floats) or "classification" (integer class indices).
    """
    for _ in range(num_steps):
        # 1. Generate X
        if batch_size is None:
            X = torch.randn(dim_x, 1)
        else:
            X = torch.randn(batch_size, dim_x, 1)
            
        # 2. Generate Y
        if target_type == "regression":
            if batch_size is None:
                Y = torch.randn(dim_y, 1)
            else:
                Y = torch.randn(batch_size, dim_y, 1)
        elif target_type == "classification":
            if batch_size is None:
                Y = torch.randint(0, dim_y, (1,)).squeeze()  # single class index
            else:
                Y = torch.randint(0, dim_y, (batch_size,))     # batch of class indices
        else:
            raise ValueError("target_type must be 'regression' or 'classification'")
            
        yield X, Y

# --- Example Usage ---

# 1. Unbatched regression data
gen = generate_data(dim_x=100, dim_y=5, batch_size=None, target_type="regression")
X, Y = next(gen)
print("Unbatched Regression -> X shape:", X.shape, "| Y shape:", Y.shape)

# 2. Batched classification data
gen_batch = generate_data(dim_x=100, dim_y=5, batch_size=32, target_type="classification")
X_b, Y_b = next(gen_batch)
print("Batched Classification -> X shape:", X_b.shape, "| Y shape:", Y_b.shape)
```

```python
import torch
from torch.utils.data import Dataset, DataLoader, random_split

# 1. Define your dataset
class RandomDataset(Dataset):
    def __init__(self, num_samples=1000, dim_x=100, dim_y=5):
        self.num_samples = num_samples
        self.dim_x = dim_x
        self.dim_y = dim_y
        
    def __len__(self):
        return self.num_samples
        
    def __getitem__(self, idx):
        X = torch.randn(self.dim_x, 1)
        Y = torch.randn(self.dim_y, 1)
        return X, Y

# 2. Instantiate dataset
full_dataset = RandomDataset(num_samples=1000, dim_x=100, dim_y=5)

# 3. Define split sizes (e.g., 70% train, 15% validation, 15% test)
train_size = int(0.70 * len(full_dataset))
val_size = int(0.15 * len(full_dataset))
test_size = len(full_dataset) - train_size - val_size

# 4. Perform split
train_dataset, val_dataset, test_dataset = random_split(
    full_dataset, 
    [train_size, val_size, test_size],
    generator=torch.Generator().manual_seed(42) # For reproducibility
)

# 5. Create DataLoaders for each split
train_loader = DataLoader(train_dataset, batch_size=32, shuffle=True)
val_loader = DataLoader(val_dataset, batch_size=32, shuffle=False)
test_loader = DataLoader(test_dataset, batch_size=32, shuffle=False)

print(f"Train samples: {len(train_dataset)}, Val samples: {len(val_dataset)}, Test samples: {len(test_dataset)}")
```

```python
import torch

# 1. Configuration
total_samples = 1000
dim_x = 100
dim_y = 5

# 2. Generate full dataset tensors
X = torch.randn(total_samples, dim_x, 1)
Y = torch.randn(total_samples, dim_y, 1)

# 3. Shuffle indices randomly for a clean distribution
indices = torch.randperm(total_samples)
X = X[indices]
Y = Y[indices]

# 4. Calculate split markers (70% Train, 15% Validation, 15% Test)
train_end = int(0.70 * total_samples)
val_end = train_end + int(0.15 * total_samples)

# 5. Extract x_train, y_train, etc.
x_train, y_train = X[:train_end], Y[:train_end]
x_val, y_val = X[train_end:val_end], Y[train_end:val_end]
x_test, y_test = X[val_end:], Y[val_end:]

# --- Verify Results ---
print(f"x_train shape: {x_train.shape} | y_train shape: {y_train.shape}")
print(f"x_val shape:   {x_val.shape}  | y_val shape:   {y_val.shape}")
print(f"x_test shape:  {x_test.shape}  | y_test shape:  {y_test.shape}")
```

```python
total_samples = 1000
dim_x = 100
num_classes = 5

# X remains the same shape: [total_samples, 100, 1]
x_data = torch.randn(total_samples, dim_x, 1)

# Y should be integer class indices: shape [total_samples] with values in range [0, 4]
y_data = torch.randint(low=0, high=num_classes, size=(total_samples,))

# Split them
train_end = int(0.70 * total_samples)
val_end = train_end + int(0.15 * total_samples)

x_train, y_train = x_data[:train_end], y_data[:train_end]
x_val, y_val = x_data[train_end:val_end], y_data[train_end:val_end]
```

```python
class L(nn.Module): 
    def __init__(self, dim1=100, dim2=10): 
        super().__init__()
        self.weights =nn.Parameter(torch.randn(size = (dim1,dim2 ),  requires_grad =True))
        self.bias = nn.Parameter(torch.randn(size = (dim1,1 ),  requires_grad =True))
    def forward(self, X): 
        return self.weights @ torch.transpose(X, 1,2) +self.bias 

#X.T = 1,100 
class NN (nn.Module): 
    def __init__(self):
        super().__init__()
        self.L1 = L(100, 1) #output 100,1 * 1, 100 = 100x100
        self.L2 = L(64, 100) # 100, 64
        self.o = L(5, 100)
    def forward(self, X:torch.Tensor):
        o1 = self.L1(X)
        o2 = self.L2(o1)
        oo = self.o(o2)
        oo = torch.sum(oo, dim =2)
 
        return oo


```

```python
model = NN()
optimizer = torch.optim.AdamW(params = model.parameters(), lr = 0.0001)
loss_fn = nn.CrossEntropyLoss()
torch.device("mps")
```

```python
EPOCHS = 1000
losses_t= []
losses_v= []
for i in range(EPOCHS): 
    model.train()
    y_pred = model(x_train)

    loss = loss_fn(y_pred, y_train)
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()
    with torch.inference_mode(): 
        losses_t.append(loss.item())
        model.eval()
        y_pred = model(x_val)

        loss = loss_fn(y_pred, y_val)
        losses_v.append(loss.item())



        
        
```

```python
import torch

# ==========================================
# 1. THE SETUP (Initialization)
# ==========================================
vocab_size = 5    # Tiny vocab: [0:'the', 1:'cat', 2:'chased', 3:'mouse', 4:'dog']
embed_dim = 4     # Each word is a 4-number vector

# The raw dense embedding table (Weights)
# We use requires_grad=True because the model needs to learn these numbers
weights = torch.randn(vocab_size, embed_dim, requires_grad=True)

# Our input sentence: "the cat chased the mouse" -> [0, 1, 2, 0, 3]
# Notice '0' (the) repeats!
token_ids = torch.tensor([0, 1, 2, 0, 3]) 


# ==========================================
# 2. THE FORWARD PASS (Gather)
# ==========================================
# GATHER EXPECTS: The index tensor must have the same number of dimensions 
# as the source tensor. `weights` is 2D, but `token_ids` is 1D.
# We have to stretch `token_ids` to match the embedding dimension so it knows 
# to grab all 4 numbers for each word.

# Stretch [5] -> [5, 4]
index_expanded = token_ids.unsqueeze(1).expand(-1, embed_dim)
# Now we pull the vectors using our stretched routing instructions
# dim=0 means "look down the rows (vocabulary) to find the vectors"
forward_output = torch.gather(weights, dim=0, index=index_expanded)

print("Forward Pass Output Shape:", forward_output.shape) 
# Result: [5, 4] (5 words, 4 dimensions each)


# ==========================================
# 3. THE BACKWARD PASS (Scatter Reduce)
# ==========================================
# Let's pretend the rest of the neural network did its math and sent back 
# a gradient tensor showing how much each of those 5 words needs to change.
# It will have the exact same shape as our forward_output [5, 4].
incoming_gradients = torch.randn(5, embed_dim) 

# SCATTER EXPECTS: A target tensor to write into. 
# We create an empty gradient tensor shaped like our massive embedding table.
sparse_grad_update = torch.zeros(vocab_size, embed_dim)

# Now we route the incoming gradients back to their original rows in the vocab.
# Because token '0' was used twice, we MUST use reduce='sum' to add its two 
# incoming gradients together, avoiding data loss!
sparse_grad_update.scatter_reduce_(
    dim=0, 
    index=index_expanded, 
    src=incoming_gradients, 
    reduce='sum'
)

print("\nSimulated Sparse Update Table:")
print(sparse_grad_update)
```
